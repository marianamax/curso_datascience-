```python
#---- ATIVIDADES DO DESAFIO EXTRA ----

# * importação dos dados;
# * compreensão do conjunto de informações;
# * preparação e tratamento dos dados;
# * verificação e tratamento de valores nulos e duplicados;
# * ajustes de tipagem;
# * organização das colunas.


# importar biblioteca e dados
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Ler o .csv
df = pd.read_csv('/content/sample_data/titanic_dataset.csv')


print("----- head -----")

# Mostrar as primeiras 5 linhas
print(df.head())

print("----- info -----")

# Mostrar as informações gerais do arquivo
df.info()

print("----- nulos -----")

# Verificar dados nulos
print(df.isnull().sum())


print("----- duplicados -----")

# verificar e exibir dados duplicados
duplicateRows = df[df.duplicated()]
print(duplicateRows)

print("----- excluir duplicados -----")

# Excluir linhas duplicadas
df = df.drop_duplicates()


print("----- tratando nulos (Age) -----")

# Substituir idades nulas com a média de idade
df['Age'] = df['Age'].fillna(df['Age'].mean())

print(df.isnull().sum())


print("----- tratando nulos (Cabin) -----")

# Informar com "Desconhecida" os campos vazios
df['Cabin'] = df['Cabin'].fillna('Desconhecida')

print(df.isnull().sum())


print("----- tratando nulos (Embarked) -----")

# Excluir as linhas em branco que o campo Embarked estiver vazio
df = df.dropna(subset=['Embarked'])

print(df.isnull().sum())


# Verificar novamente se há nulos
print(df.isnull().sum())


print("----- tipos de dados -----")

# Verificar tipos de dados
print(df.dtypes)

print("----- convertendo idade -----")

# Alterando a Idade para número inteiro
df['Age'] = df['Age'].astype(int)

# Verificar a mudança
print(df['Age'].head())


print("----- ordenação -----")

# Ordenação por sexo - primeiro feminino depois masculino
df_ordenado = df.sort_values(by='Sex')
print(df_ordenado.head())


print("----- agrupamento -----")

# groupby
df.groupby(['Sex'])['Survived'].sum()

```

```python


# Mostrar os dados no gráfico
# importar biblioteca e dados
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset('titanic')

df_por_sexo = titanic.groupby('sex')['survived'].sum().reset_index()

# Criando um gráfico de barras
plt.figure(figsize=(8,6))

sns.barplot(data=df_por_sexo, x='sex', y='survived')

plt.show()
```
